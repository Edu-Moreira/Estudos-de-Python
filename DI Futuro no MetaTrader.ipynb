```python
# Define o intervalo de datas para o histrico
start_date = datetime(2024, 6, 1)
end_date = datetime.now()

# Define o ativo e o perodo de tempo
symbol = "VALE3"
timeframe = mt.TIMEFRAME_D1  # Dirio

# Pega o histrico de preos
rates = mt.copy_rates_range(symbol, timeframe, start_date, end_date)

# Converte para DataFrame do pandas
rates_frame = pd.DataFrame(rates)

# Converte o timestamp para datetime
rates_frame['time'] = pd.to_datetime(rates_frame['time'], unit='s')

print(rates_frame.tail())
```

```python
colecao_di = []

for symbol in ticker:
    # Pega o histrico de preos
    rates = mt.copy_rates_range(symbol, timeframe, start_date, end_date)
    # Converte para DataFrame do pandas
    rates_frame = pd.DataFrame(rates)
    # Converte o timestamp para datetime
    rates_frame['time'] = pd.to_datetime(rates_frame['time'], unit='s')
    
    fech_di = pd.DataFrame(rates_frame.close.tail(5))
    fech_di.columns=[symbol]
    # fech_di.reset_index(inplace=True)
    fech_di.index = rates_frame.time.tail(5)
    # fech_di.drop(columns= 'index',inplace=True)
       
    colecao_di.append(fech_di)

```

```python
dfs_non_empty = [df for df in colecao_di if not df.empty]


dfs_non_empty
historico_di = pd.concat(dfs_non_empty,axis=1)
```

```python
# historico_di = historico_di.dropna(thresh=10)

historico_di.dropna(how='all',axis=1,inplace=True)

historico_di = historico_di.ffill()


if len(historico_di) < 3:
    print('Erro Histrico Menor que 3 dias.......Favor Verificar')
else:
    historico_di = historico_di.tail(3)
```

```python
historico_di.ffill()
```

```python
cadastro_dis = {ticker: lista_vencimentos[ticker] for ticker in historico_di.columns}

historico_di_invertido = historico_di.T
historico_di_invertido['Ticker'] = historico_di_invertido.index
historico_di_invertido
historico_di_invertido['Vencimento'] = historico_di_invertido['Ticker'].replace(cadastro_dis)

historico_di_invertido.columns = ['D-3', 'D-2', 'D-1', 'Ticker', 'Vencimento']

historico_di_invertido.index = historico_di_invertido['Vencimento']

historico_di_invertido[['D-3','D-1']].plot(figsize = (21,4))
```
